```python
def deriv(theta, t): # return derivatives of the array theta 
    g = 9.8 #m/s^2
    l = 0.5 #m
    return array([ theta[1], -(g/l)*sin(theta[0]) ])
```

```python
theta_app = lambda t,th0: th0*cos(sqrt(g/l)*t)

th0 = 0
theta_init = array([th0, omega0]) # initial values 
theta = odeint(deriv, theta_init, t)

th1 = pi/4 
theta_init = array([th1, omega0]) # initial values 
theta = odeint(deriv, theta_init, t)

th2 = pi/2 
theta_init = array([th2, omega0]) # initial values 
theta = odeint(deriv, theta_init, t)

th3 = 3*pi/4 
theta_init = array([th3, omega0]) # initial values 
theta = odeint(deriv, theta_init, t)

th4 = pi 
theta_init = array([th4, omega0]) # initial values 
theta = odeint(deriv, theta_init, t)

plot(t, theta_app(t,th0),'b-', t, theta[:,0], 'r--') #0
plot(t, theta_app(t,th1),'r-', t, theta[:,0], 'r--') #pi/4
plot(t, theta_app(t,th2),'g-', t, theta[:,0], 'r--') #pi/2
plot(t, theta_app(t,th3),'m-', t, theta[:,0], 'r--') #3pi/4
plot(t, theta_app(t,th4),'k--', t, theta[:,0], 'r--') #pi
show()
```

The Period of the Pendulum increases as the initial angle is increased from pi/4 to 3*pi/4. The Pendulum is stationary (has no trajectory) when theta is equal to the values of 0 and pi.
