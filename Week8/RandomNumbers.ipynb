##Example 1: Analytic Inverse CDF Transform

```python
x = arange(0., pi+0.001, 0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.

plot(x,f(x),label='True distribution', lw=3)
plot(x,F(x), label='Cumulative function', lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)

#sampled x values 
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label="Sample distribution")
plot(x,f(x),label='True distribution', lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

##Exercise 1. Exponential function - analytic

```python
#Parts A/B
f = lambda x,lm: lm*e**(-lm*x)
F = lambda x,lm: 1 - e**(-l*x) #integral is -e^(-l * x)
Finv = lambda x,lm: log(1 - x)/(lm) 

x = arange(0., 10+0.001, 0.01)
lm = 1

plot(x,f(x,lm),label='True distribution', lw=3)
plot(x,F(x,lm),label='Cumulative function', lw=3)

legend(loc=0)
show()
```

```python
#Part C
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda x,lm: -log(lm-x)/(lm) 

#sampled x values 
xsamples = Finv(z,lm);

#how do they compare to the original?
hist(xsamples,50,normed=True,label="Sample distribution")
plot(x,f(x,lm),label='True distribution', lw=3,color='red')

legend(loc=0)
show()
```

###Exercise 2

```python
x = arange(0., 10+0.001, 0.01)
lm = 1
f = lambda x,lm: lm*e**(-lm*x)
N = 10000
z = numpy.random.random_sample(N);

farray = f(x, lm)    


W = cumsum(farray)
X = sum(farray)
normalize = W/X

indexvalues = [] 
for a in z:
    indexvalues = append(indexvalues, searchsorted(normalize,a))
    
emptyarray = []
for j in indexvalues:
    emptyarray = append(emptyarray,x[j])
```

```python
hist(emptyarray,50,normed=True,label="Sample distribution")
plot(x,f(x,lm),label='True distribution', lw=3,color='red')

legend(loc=0)
show()
```
