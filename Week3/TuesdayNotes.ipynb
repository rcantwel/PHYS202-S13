Create a very simple plot

```python
import numpy as np
t = arange(0.0 , 1.0+0.01 , 0.01)
c = cos(2*2*pi*t)
s = sin(2*2*pi*t)
```

```python
print pi
```

```python
plot(t,c, color = "blue", linewidth = 2.5,linestyle= "--", label= "$cos(4{\pi}t)$")
plot(t,s, color = "red",linewidth = 2.5, linestyle= ":", label = "$sin(4{\pi}t)$")
xlabel(' time (s) ')
ylabel('voltage (mV)')
#Add title for convenience 
title('A simple plot', fontsize=20)
grid(True)
legend(loc = "lower left")
savefig('simple_plot.pdf')
```
