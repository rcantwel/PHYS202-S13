#3-D Plotting and Functions

###Electric Dipole Potential

$V(x,y) = \cfrac{kq}{\sqrt{x^2 +(y-d)^2}} - \cfrac{kq}{\sqrt{x^2 +(y+d)^2}} $

```python
#Make the x-y grid to cover the plane -2<x<2 and -2<y<2
dx = .06
dy = .06 
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy)) 
```

```python
#Make the function, V(x,y) 
k=1; q=1;d=1 
V=(k*q/(x**2+(y-d)**2)**(1/2.)) - (k*q/(x**2+(y+d)**2)**(1/2.)) 
```

```python
#Convenient way to create a 3D axis 
from mpl_toolkits.mplot3d import Axes3D

fig = figure() 
ax = fig.gca(projection = '3d') 
p = ax.plot_surface( x, y, V, rstride = 1, cstride = 1, linewidth = 0, cmap=cm.coolwarm)


cb = fig.colorbar(p, shrink = 0.5)



cset = ax.contour(x, y, V, zdir = 'z', ofset = -40., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'x', ofset = -2., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'y', ofset = -2., cmap = cm.coolwarm)
```

```python
help(Axes3D.plot_surface)
```

```python
levels = array([8,4,2,1,0.5, 0.25, -0.25, -0.50,-1,-2,-4,-8]) 
contour(x, y, V, levels, linewidths = 2, cmap = cm.coolwarm)
show ()
```
